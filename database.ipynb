```python
# get connexion to database
connection = pymysql.connect(host='localhost', 
                             user=USERNAME, 
                             password=PASSWORD, 
                             db=DATABASE)

cursor = connection.cursor()
```

```python
DB_URL = 'mysql+pymysql://{}:{}@{}:{}/{}?charset=utf8'.format(USERNAME, PASSWORD, HOST, PORT, DATABASE)

engine = create_engine(DB_URL)
```

```python
# load csv
tags = pd.read_csv('data/tags.csv')
# add tags csv to database
tags.to_sql('tags', con=engine, if_exists='append', chunksize=1000)
```

```python
# load csv
books = pd.read_csv('data/books.csv')
# add books csv to database
books.to_sql('books', con=engine, if_exists='append', chunksize=1000)
```

```python
# load csv
books_with_cat = pd.read_csv('data/books_with_cat.csv')
# add books with tags csv to database
books_with_cat.to_sql('books_with_cat', con=engine, if_exists='append', chunksize=1000)
```

```python
# load csv
ratings = pd.read_csv('data/ratings.csv')
# add ratings csv to database
ratings.to_sql('ratings', con=engine, if_exists='append', chunksize=1000)
```
